```python
train_data=pd.read_csv('train.csv')
test_data=pd.read_csv('test.csv')
frames=[train_data.drop(['ID','y'],axis=1),test_data.drop('ID',axis=1)]
dataset=pd.concat(frames)
```

```python
x0=pd.get_dummies(dataset['X0'],drop_first=True)
x1=pd.get_dummies(dataset['X1'],drop_first=True)
x2=pd.get_dummies(dataset['X2'],drop_first=True)
x3=pd.get_dummies(dataset['X3'],drop_first=True)
x4=pd.get_dummies(dataset['X4'],drop_first=True)
x5=pd.get_dummies(dataset['X5'],drop_first=True)
x6=pd.get_dummies(dataset['X6'],drop_first=True)
x8=pd.get_dummies(dataset['X8'],drop_first=True)
```

```python
dataset.drop(['X0','X1','X2','X3','X4','X5','X6','X8'],axis=1,inplace=True)
```

```python
dataset = pd.concat([dataset,x0,x1,x2,x3,x4,x5,x6,x8],axis=1)
```

```python
dataset.shape
```

```python
dataset=dataset.loc[:,~dataset.columns.duplicated()]
```

```python
dataset.shape
```

```python
# Import Independent Component Analysis Algorithm
from sklearn.decomposition import FastICA

# Initialize the algorithm and set the number of PC's
ica = FastICA(n_components=3)

# Fit and transform the model to data. It returns a list of independent components 
data=ica.fit_transform(dataset)
```

```python
data
```

```python
X_train=dataset.iloc[:4209,:]
X_test=dataset.iloc[4209:,:]
```

```python
X_train.shape
```

```python
train_data.shape
```

```python
import xgboost
classifier=xgboost.XGBRegressor()





# Set up the random search with 4-fold cross validation

classifier.fit(X_train,train_data['y'])

```

```python
y_pred=classifier.predict(X_test)
```

```python
y_pred=pd.DataFrame(y_pred)
```
